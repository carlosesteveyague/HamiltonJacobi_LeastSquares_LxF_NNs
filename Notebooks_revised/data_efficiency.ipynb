# Comparison data efficiency

## Fixed uniformly random points

We consider 2D Eikonal equation in the suqare $(-3,3)^2$ and different sampling strategies during the training.

We consdier the following sequence of values for $\delta$ and $\alpha$:
$$
\delta = [0.75, 0.5, 0.3, 0.1, 0.05] \quad  \text{and} \quad
\alpha = [2.5, 2, 1.5, 1, 0.5].
$$

```python
import torch.optim as optim
import matplotlib.pyplot as plt
import torch
import sys
# setting path 
sys.path.append('../')
from NeuralNetworks.NNs import FCFF_3L
from PointSampling.Cube import data_gen_cube, get_unif_grid
from visualization.plots_cube import plot_2d_proj, plot_level_set_cube 
from visualization.cube_training import plot_2d_proj_w
from Hamiltonians.Eikonal_LxF import Eikonal_sq_LF_multiD 
from Training.training_finite_sum import train_finite_sum
from error_test.cube_error import error_cube 

dim = 2


side_length = 6.

domain = data_gen_cube(side_length, dim)
#Right-hand-side of the PDE
def f(X): return 1
#Boundary condition
def g(X): return 0
delta_list = [.75, .5, .3, .1, .05]
alpha_list = [2.5, 2., 1.5, 1., .5]
rounds = len(delta_list)
training_params = {
    'numerical_scheme': Eikonal_sq_LF_multiD,
    
    'f': f, 
    'g': g, 
    'c': None,
    
    'beta': 0., ## parameter for the +u_i term 
    
    'lambda': 1. #weight parameter for the boundary loss
}
```

## Training with fixed data points

First we consider the setting in which the collocation points randomly sampled at the beginning,
and then are re-used throughout the training.

As training data, we use $80$ collocation points, sampled uniformly at random from $(-3,3)^2$, and $40$ boundary points sampled uniformly at random from the boundary.

We carry out $500$ epochs of SGD, with batch size 20. This means that every collocation point and boundary point in the dataset are used 500 times in each training round.

After each training round (before reducing $\delta$ and $\alpha$), we plot:

1. The evolution of the error throughout the epochs of each training round,
2. The numerical solution provided by the NN,
3. The characteristic function associated to the set
$$
\left\{  x\in \Omega \ : \quad \left( \widehat{H} (x, D_\delta^+ u(x) , D_\delta^- u(x)) - f(x) \right)^2 < 10^{-2} \right\},
$$
and the collocation points (in red).

```python
# We initialise the NN
NN1 = FCFF_3L([dim,20,20])
training_params['optimizer'] = optim.SGD(NN1.parameters(), lr = .02, momentum = .2)

## Here we sample the collocation points and the boundary points.
int_points = domain.rand_int_points(0., 80)
bound_points = domain.rand_bound_points(40)

training_params['epochs'] = 500
training_params['batch_size'] = 20

MSE_history = torch.zeros(rounds)
L_inf_error_history = torch.zeros(rounds) 

for i in range(rounds):
    
    training_params['alpha'] = alpha_list[i]
    training_params['delta'] = delta_list[i]

    total_loss, PDE_loss, boundary_loss = train_finite_sum(NN1, int_points, bound_points, training_params)

    MC_points = int(1e5) # Number of grid points for comparison with the ground truth
    MSE, L_inf = error_cube(NN1, side_length, MC_points)
    MSE_history[i] = MSE
    L_inf_error_history[i] = L_inf
    
    epochs =  training_params['epochs']
    plt.plot(torch.arange(epochs)+1, total_loss)
    plt.plot(torch.arange(epochs)+1, PDE_loss)
    plt.plot(torch.arange(epochs)+1, boundary_loss)
    plt.legend(labels=['Total loss', 'PDE loss', 'BC loss'])
    plt.show()
    
    X_axis = 0
    Y_axis = 1
    n_grid = 100
    
    plot_2d_proj(X_axis, Y_axis, NN1, n_grid, side_length)
    plot_2d_proj_w(X_axis, Y_axis, NN1, n_grid, side_length, training_params,int_points, tol = 1e-2)
    
    print('Mean square error:', MSE)
    print('L-infinity error:', L_inf)
```

## Resampling after each training round

Here we re-sample the data points after each training round, i.e. before reducing $\alpha$ and $\delta$.

```python
# We define the NN again, so the parameters are initialised
NN2 = FCFF_3L([dim,20,20])
training_params['optimizer'] = optim.SGD(NN2.parameters(), lr = .02, momentum = .2)

MSE_history = torch.zeros(rounds)
L_inf_error_history = torch.zeros(rounds) 


for i in range(rounds):
    ## Here we sample the collocation points and the boundary points.
    int_points = domain.rand_int_points(0., 80)
    bound_points = domain.rand_bound_points(40)
    
    training_params['alpha'] = alpha_list[i]
    training_params['delta'] = delta_list[i]
    
    total_loss, PDE_loss, boundary_loss = train_finite_sum(NN2, int_points, bound_points, training_params)
    
    MC_points = int(1e5) # Number of grid points for comparison with the ground truth
    MSE, L_inf = error_cube(NN2, side_length, MC_points)
    MSE_history[i] = MSE
    L_inf_error_history[i] = L_inf
   
    epochs =  training_params['epochs']
    plt.plot(torch.arange(epochs)+1, total_loss)
    plt.plot(torch.arange(epochs)+1, PDE_loss)
    plt.plot(torch.arange(epochs)+1, boundary_loss)
    plt.legend(labels=['Total loss', 'PDE loss', 'BC loss'])
    plt.show()
   
    X_axis = 0
    Y_axis = 1
    n_grid = 100
    plot_2d_proj(X_axis, Y_axis, NN2, n_grid, side_length)
    plot_2d_proj_w(X_axis, Y_axis, NN2, n_grid, side_length, training_params, int_points, tol = 1e-2)
    
    print('Mean square error:', MSE)
    print('L-infinity error:', L_inf)
```

## Re-sampling at every iteration

Here we re-sample the collocation points and the boundary points at every gradient descent iteration. We use $12$ collocation points and $8$ boundary points and perform $2500$ iterations of GD. In this case, the sampled points are used only once.

In this case, we do not plot the collocation points, since in every training round, we are using a total of $12 ∗ 2500 = 30000$ different points in the interior.

Note that in the other experiments, for a dataset with $80 + 40 = 120$, we were doing $500$ epochs with batches of size $20$. This corresponds to $2500$ iterations, with $20$ points in each iteration.

```python
from Training.training import train

# We define the NN again, so the parameters are initialised
NN4 = FCFF_3L([dim,20,20])
training_params['optimizer'] = optim.SGD(NN4.parameters(), lr = .02, momentum = .2)

training_params['n_coloc_points'] = 20
training_params['n_boundary_points'] = 8
training_params['num_iterations'] = 2500

for i in range(rounds):
    
    training_params['alpha'] = alpha_list[i]
    training_params['delta'] = delta_list[i]
    
    total_loss, PDE_loss, boundary_loss = train(NN4, domain, training_params)
    
    plt.plot(total_loss)
    plt.plot(PDE_loss)
    plt.plot(boundary_loss)
    plt.legend(labels=['Total loss', 'PDE loss', 'BC loss'])
    plt.show()

    MC_points = int(1e5) # Number of grid points for comparison with the ground truth
    MSE, L_inf = error_cube(NN4, side_length, MC_points)
    MSE_history[i] = MSE
    L_inf_error_history[i] = L_inf
    
    X_axis = 0
    Y_axis = 1
    n_grid = 100
    plot_2d_proj(X_axis, Y_axis, NN4, n_grid, side_length)
    plot_2d_proj_w(X_axis, Y_axis, NN4, n_grid, side_length, training_params, tol = 1e-2)
    print('Mean square error:', MSE)
    print('L-infinity error:', L_inf)
```
